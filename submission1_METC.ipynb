```python
import pandas as pd
from sklearn.datasets import load_diabetes
# 1. Memuat dataset diabetes standar dari scikit-learn
diabetes_data = load_diabetes(as_frame=True)
df = diabetes_data.frame

# 2. Menyelaraskan nama kolom agar persis seperti versi Azure Open Dataset di modul
# Versi scikit-learn aslinya menggunakan huruf kecil, kita ubah ke huruf besar
df.columns = [col.upper() for col in df.columns]

# Versi scikit-learn mengasumsikan kolom target bernama 'TARGET', kita ubah menjadi 'Y'
df = df.rename(columns={'TARGET': 'Y'})

# 3. Menghapus skala (unscaling) data agar angkanya kembali menjadi angka riil (seperti AGE 59, BMI 32.1)
# Karena scikit-learn secara bawaan sudah men-scale datanya.
raw_data = load_diabetes()
df_raw = pd.DataFrame(raw_data.data, columns=[col.upper() for col in raw_data.feature_names])

# Mengembalikan nilai asli ke kolom numerik agar sama dengan modul Dicoding Anda
df['AGE'] = (df_raw['AGE'] * 100 + 48.5).astype(int) # Pendekatan nilai rata-rata umur
df['BMI'] = (df_raw['BMI'] * 100 + 26.3).round(1)

df_clean = df
# Menampilkan 5 data teratas
df_clean.head()
```

```python
import pandas as pd
from sklearn.datasets import load_diabetes

# 1. Load data diabetes dari scikit-learn
diabetes = load_diabetes()
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)

# Memasukkan target asli (variabel kontinu Y) untuk REGRESI
df['Y'] = diabetes.target

# 2. Membuat kolom biner 'Risk' untuk KLASIFIKASI
df['Risk'] = (df['Y'] > 211.5).astype(int)

# Kolom 'Y' TIDAK dihapus agar bisa dipakai latihan model regresi
print(df.head())
```

```python
from sklearn.model_selection import train_test_split

# Fitur menggunakan nama huruf kecil sesuai data scikit-learn
X = df[['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6']].values
y = df['Y'].values # Target regresi menggunakan Y

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
import mlflow
experiment_name = "diabetes-regression"
mlflow.set_experiment(experiment_name)
```

```python
from sklearn.linear_model import LinearRegression
    
with mlflow.start_run():
   mlflow.autolog()
    
   model = LinearRegression()
   model.fit(X_train, y_train)
```
