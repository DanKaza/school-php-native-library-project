```python
import pandas as pd
from sklearn.datasets import load_diabetes

# 1. Memuat dataset diabetes standar dari scikit-learn
diabetes_data = load_diabetes(as_frame=True)
df = diabetes_data.frame

# 2. Menyelaraskan nama kolom agar persis seperti versi Azure Open Dataset di modul
# Versi scikit-learn aslinya menggunakan huruf kecil, kita ubah ke huruf besar
df.columns = [col.upper() for col in df.columns]

# Versi scikit-learn mengasumsikan kolom target bernama 'TARGET', kita ubah menjadi 'Y'
df = df.rename(columns={'TARGET': 'Y'})

# 3. Menghapus skala (unscaling) data agar angkanya kembali menjadi angka riil (seperti AGE 59, BMI 32.1)
# Karena scikit-learn secara bawaan sudah men-scale datanya.
raw_data = load_diabetes()
df_raw = pd.DataFrame(raw_data.data, columns=[col.upper() for col in raw_data.feature_names])

# Mengembalikan nilai asli ke kolom numerik agar sama dengan modul Dicoding Anda
df['AGE'] = (df_raw['AGE'] * 100 + 48.5).astype(int) # Pendekatan nilai rata-rata umur
df['BMI'] = (df_raw['BMI'] * 100 + 26.3).round(1)

# Menampilkan 5 data teratas
df.head()
```

```python
# Display the number of rows and columns in the dataset
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# Display the data types of each column
print("\nData types of columns:")
print(df.dtypes)
```

```python
missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
    
# Calculate the mean, median of the BMI variable
mean = df['BMI'].mean()
median = df['BMI'].median()
   
# Histogram of the BMI variable
plt.figure(figsize=(8, 6))
plt.hist(df['BMI'], bins=20, color='skyblue', edgecolor='black')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
    
# Add lines for the mean and median
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label='Median')
    
# Add a legend
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of BMI vs. Target variable
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Y', data=df)
plt.title('BMI vs. Target variable')
plt.xlabel('BMI')
plt.ylabel('Target')
plt.show()
```

```python
import numpy as np

# Mengubah nilai kolom SEX berdasarkan dataset asli scikit-learn
# Nilai negatif menjadi 'Male', nilai positif menjadi 'Female'
df['SEX'] = np.where(df['SEX'] < 0, 'Male', 'Female')

# Cek apakah datanya sudah benar-benar berubah menjadi teks
print(df['SEX'].unique())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
    
fig, ax = plt.subplots(figsize=(7, 5))
    
# Replace numeric values with labels
df['SEX'] = df['SEX'].replace({1: 'Male', 2: 'Female'})
    
sns.boxplot(x='SEX', y='BP', data=df, ax=ax)
ax.set_title('Blood pressure across Gender')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
    
# Calculate average BP and BMI by SEX
avg_values = df.groupby('SEX')[['BP', 'BMI']].mean()
    
# Bar chart of the average BP and BMI by SEX
ax = avg_values.plot(kind='bar', figsize=(15, 6), edgecolor='black')
    
# Add title and labels
plt.title('Avg. Blood Pressure and BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('Average')
    
# Display actual numbers on the bar chart
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 10), 
                textcoords = 'offset points')
    
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
    
plt.figure(figsize=(10, 6))
sns.lineplot(x='AGE', y='BMI', data=df, errorbar=None)
plt.title('BMI over Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```
