```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the Penguins dataset from seaborn
penguins = sns.load_dataset('penguins')

# Create a binomial variable for 'species'
penguins['is_adelie'] = np.where(penguins['species'] == 'Adelie', 1, 0)

# Plot the distribution of 'is_adelie'
sns.histplot(data=penguins, x='is_adelie', bins=2, discrete=True)
plt.title('Binomial Distribution of Species')
plt.xticks([0, 1], ['Not Adelie', 'Adelie'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Mengambil dataset bawaan seaborn (tidak perlu unduh file luar)
data_restoran = sns.load_dataset('tips')

# 2. Mengubah nama kolom ke bahasa Indonesia agar lebih mudah dipahami
data_restoran = data_restoran.rename(columns={
    'total_bill': 'Total_Tagihan',
    'tip': 'Tip_Pelayan',
    'sex': 'Gender_Pelanggan',
    'smoker': 'Perokok',
    'day': 'Hari',
    'time': 'Waktu_Makan',
    'size': 'Jumlah_Orang'
})

# 3. Membuat grafik Scatter Plot (titik-titik korelasi)
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=data_restoran, 
    x='Total_Tagihan', 
    y='Tip_Pelayan', 
    hue='Gender_Pelanggan', # Membedakan warna titik berdasarkan gender
    style='Perokok'          # Membedakan bentuk titik berdasarkan perokok/tidak
)

# 4. Merapikan tampilan grafik
plt.title('Hubungan Total Tagihan dan Tip yang Diterima Pelayan')
plt.xlabel('Total Tagihan Makan (USD)')
plt.ylabel('Tip untuk Pelayan (USD)')
plt.grid(True, linestyle='--', alpha=0.5)

# 5. Menampilkan grafik
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Mengambil dataset resmi 'diamonds'
data_berlian = sns.load_dataset('diamonds')

# 2. Mengubah nama kolom yang SESUAI dengan isi data berlian
data_berlian = data_berlian.rename(columns={
    'carat': 'Berat_Karat',
    'price': 'Harga_Berlian'
})

# Kita ambil 500 sampel data saja agar grafik regresinya ringan dan cepat muncul
data_sampel = data_berlian.sample(n=500, random_state=42)

# 3. Membuat grafik regresi linear berdasarkan nama kolom yang baru
sns.lmplot(
    data=data_sampel, 
    x='Berat_Karat', 
    y='Harga_Berlian',
    height=5, 
    aspect=1.5,
    line_kws={'color': 'red'} 
)

# 4. Merapikan teks keterangan grafik
plt.title('Model Regresi: Prediksi Harga Berlian Berdasarkan Karat')
plt.xlabel('Berat Berlian (Karat)')
plt.ylabel('Harga Berlian (USD)')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = pd.read_csv('https://raw.githubusercontent.com/MicrosoftLearning/dp-data/main/penguins.csv')

# Remove missing values
penguins = penguins.dropna()

# Prepare the data and target
X = penguins.drop('Species', axis=1)
y = penguins['Species']

# Initialize and apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the data
plt.figure(figsize=(8, 6))
for color, target in zip(['navy', 'turquoise', 'darkorange'], penguins['Species'].unique()):
 plt.scatter(X_pca[y == target, 0], X_pca[y == target, 1], color=color, alpha=.8, lw=2,
 label=target)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Penguins dataset')
plt.show()
```
