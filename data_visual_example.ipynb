```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the Penguins dataset from seaborn
penguins = sns.load_dataset('penguins')

# Create a binomial variable for 'species'
penguins['is_adelie'] = np.where(penguins['species'] == 'Adelie', 1, 0)

# Plot the distribution of 'is_adelie'
sns.histplot(data=penguins, x='is_adelie', bins=2, discrete=True)
plt.title('Binomial Distribution of Species')
plt.xticks([0, 1], ['Not Adelie', 'Adelie'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Mengambil dataset bawaan seaborn (tidak perlu unduh file luar)
data_restoran = sns.load_dataset('tips')

# 2. Mengubah nama kolom ke bahasa Indonesia agar lebih mudah dipahami
data_restoran = data_restoran.rename(columns={
    'total_bill': 'Total_Tagihan',
    'tip': 'Tip_Pelayan',
    'sex': 'Gender_Pelanggan',
    'smoker': 'Perokok',
    'day': 'Hari',
    'time': 'Waktu_Makan',
    'size': 'Jumlah_Orang'
})

# 3. Membuat grafik Scatter Plot (titik-titik korelasi)
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=data_restoran, 
    x='Total_Tagihan', 
    y='Tip_Pelayan', 
    hue='Gender_Pelanggan', # Membedakan warna titik berdasarkan gender
    style='Perokok'          # Membedakan bentuk titik berdasarkan perokok/tidak
)

# 4. Merapikan tampilan grafik
plt.title('Hubungan Total Tagihan dan Tip yang Diterima Pelayan')
plt.xlabel('Total Tagihan Makan (USD)')
plt.ylabel('Tip untuk Pelayan (USD)')
plt.grid(True, linestyle='--', alpha=0.5)

# 5. Menampilkan grafik
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Ambil data dan ubah nama kolom seperti sebelumnya
data_restoran = sns.load_dataset('tips')
data_restoran = data_restoran.rename(columns={
    'total_bill': 'Total_Tagihan',
    'tip': 'Tip_Pelayan'
})

# 2. Membuat grafik Scatter Plot SEKALIGUS Garis Regresi Linear
sns.lmplot(
    data=data_restoran, 
    x='Total_Tagihan', 
    y='Tip_Pelayan',
    height=6, 
    aspect=1.5,
    line_kws={'color': 'red'} # Mengubah warna garis regresi menjadi merah agar jelas
)

# 3. Merapikan teks
plt.title('Model Regresi Linear: Prediksi Tip Berdasarkan Total Tagihan')
plt.xlabel('Total Tagihan Makan (USD)')
plt.ylabel('Tip untuk Pelayan (USD)')

plt.show()
```
